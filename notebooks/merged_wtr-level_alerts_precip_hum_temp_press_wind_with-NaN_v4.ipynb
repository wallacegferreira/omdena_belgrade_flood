# Cleaning and renaming columns

```python
df_hum.columns = ['0', 'date', 'belgrade_humidity_pct']
df_hum = df_hum.drop(columns=['0'])
df_hum.info()
```

```python
df_press.columns = ['0', 'date', 'belgrade_pressure_hg']
df_press = df_press.drop(columns=['0'])
df_press.info()
```

```python
#convert milibars to inHg
df_press['belgrade_pressure_hg'] = df_press['belgrade_pressure_hg']*0.02953
```

```python
df_temp.columns = ['0', 'date', 'belgrade_temperature_c']
df_temp = df_temp.drop(columns=['0'])
df_temp.info()
```

```python
df_wind.columns = ['0', 'date', 'belgrade_windspeed_kph']
df_wind = df_wind.drop(columns=['0'])
df_wind.info()
```

```python
df_pth_2000.columns =['date', 'belgrade_pressure_hg', 'belgrade_temperature_c', 'belgrade_humidity_pct']
df_pth_2000.info()
```

```python
#convert milibars to inHg
df_pth_2000['belgrade_pressure_hg'] = df_pth_2000['belgrade_pressure_hg']*0.02953
```

```python
df_wind_2000.columns =['date', 'belgrade_windspeed_kph']
df_wind_2000.info()
```

# Merge data

## Merge pressure, temperature and humidity

```python
df_pth_2010 = df_press.copy()
df_pth_2010 = df_pth_2010.merge(df_temp,on='date',how='left')
df_pth_2010 = df_pth_2010.merge(df_hum,on='date',how='left')
df_pth_2010.info()
```

```python
df_pth = pd.concat([df_pth_2000,df_pth_2010],axis=0)
df_pth.info()
```

```python
df_pth.head()
```

## Merge windspeed

```python
df_wind = pd.concat([df_wind_2000,df_wind],axis=0)
df_wind.head()
```

```python
df_wind.info()
```

## Merge remaining features to the main dataframe

```python
df = df.merge(df_pth,on='date',how='left')
df = df.merge(df_wind,on='date',how='left')
df.info()
```

```python
df.head()
```

```python
df.columns
```

```python
df.columns =['date', 'belgrade_water_level_cm', 'belgrade_flood_alert',
       'pancevo_water_level_cm', 'pancevo_flood_alert', 'zemun_water_level_cm',
       'zemun_flood_alert', 'belgrade_precipitation_mm',
       'belgrade_pressure_hg', 'belgrade_temperature_c',
       'belgrade_humidity_pct', 'belgrade_windspeed_kph']
```

```python
df.head()
```

```python
df.to_csv('merged_wtr-level_alerts_precip_hum_temp_press_wind_with-NaN_v4.csv',index=False)
```
